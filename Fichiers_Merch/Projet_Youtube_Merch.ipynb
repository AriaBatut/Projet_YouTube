### Librairies importés

Nous allons tout d'abord importer les différentes librairies utiles dans ce projet. Pandas pour les dataframes, BeautifulSoup pour le code source html, et Request et urlopen pour accéder aux différents sites Internet.

```python
import pandas as pd
from bs4 import BeautifulSoup as bs
from urllib.request import Request, urlopen
```

On va maintenant ouvrir le fichier csv avec toutes les données sur les chaînes Youtube.

```python
data = pd.read_csv("data.csv")
data.shape
```

Regardons les données.

```python
data.head
```

Maintenant regardons les données utiles. La colonne "yt_user_id" est utile pour accéder aux chaînes Youtube. Les données correspondent à une partie spécifique de l'URL.

```python
data_fin=data.yt_user_id
print(data_fin)
```

Nous ajoutons maintenant le reste de l'URL, ce qui va le rendre utilisable par le futur.

```python
j=0
urls = [] #Pour pouvoir stocker les URLs des chaînes
for i in data_fin: #i correpond ici à la partie unique de l'URL
    i = "https://www.youtube.com/channel/UC" + i + "/about"
    urls.append(i)
    j=j+1
print(urls)
```

Nous avons maintenant un tableau contenant tous les URLs des chaînes. Nous allons donc parcourir ces URLs, afin de récupérer les liens liés à la chaîne Youtube. On va effectuer un tri des liens, c'est à dire que nous allons enlever les liens pouvant amener à des réseaux sociaux, ou à des sites de collecte de dons (de crowdfunding).

```python
urls_merch = []
for url in urls:
    #On ouvre les pages Internet à partir des liens stockés au dessus
    req = Request(url , headers={'User-Agent' : 'Mozilla'})
    webpage = urlopen(req).read()
    page_soup = bs(webpage, "html.parser") #On stocke le code HTML dans une variable afin de pouvoir le parcourir
    one_a_tag = page_soup.findAll('script') #On récupère tout le code compris entre des balises <script>
    one_a_tagS = str(one_a_tag)
    one_a_tag1 = one_a_tagS.split("www") #On recherche à créer des tableaux dont la première case sera un lien internet
    one_a_tagT=[]
    j = len(one_a_tag1)
    for i in range(j): #On va créer une boucle afin de pouvoir récuperer tous les liens disponibles, et les trier
        if  ("tiktok" not in one_a_tag1[i]) and ("google" not in one_a_tag1[i]) and ( "instagram" not in one_a_tag1[i]) and ("facebook" not in one_a_tag1[i]) and ("twitch" not in one_a_tag1[i]) and ("twitter" not in one_a_tag1[i]) and ("snapchat" not in one_a_tag1[i]) and ("utip" not in one_a_tag1[i]) and ("tipee" not in one_a_tag1[i]) and ("%2" not in one_a_tag1[i]) and ("smarturl" not in one_a_tag1[i]) and ((".fr"  in one_a_tag1[i]) or (".com" in one_a_tag1[i])) and i!=0:
            if "youtube" not in one_a_tag1[i]:
                one_a_tagT.append(one_a_tag1[i]) #On enregistre dans un tableau le code comprenant l'URL
    if(len(one_a_tagT)>2):
        one_a_tag2 = one_a_tagT[1].split("\"") #On récupère la partie du lien qui nous intéresse
        one_a_tag3 = one_a_tag2[0]
        if '\\n' in one_a_tag3: #Certains liens sont mal écrits et contiennent des retours à la ligne. On utilise ce test afin de les enlever
            one_a_tag4 = one_a_tag3.split()
            one_a_tag5 = one_a_tag4[0].split('\\n')
            one_a_tag6 = "https://www"+one_a_tag5[0]
            urls_merch.append(one_a_tag6)
            print(one_a_tag6)
        else: #On traite ensuite les liens normaux ici
            one_a_tag3 = "https://www"+one_a_tag3
            urls_merch.append(one_a_tag3)
            print(one_a_tag3)
        
    
```
